```python
lst = [ 1,2,3, 'hello', 'c', [6,8]]

print(lst)

#Append
lst.append([2,3,4])
print(lst)


#extend
lst.extend([6,7,9])
print(lst)


#insert()
#lst.insert(index, value)

lst.insert(2,'value')
print(lst)
```

```python
lst = [1,2,3,'name','rew', 99]
print(lst)

#append
#treat every value as a single value
lst.append(['yesh',2,7,8])
print(lst)

#extend
#treat every value sapretely
lst.extend(['hai sharthi'])
print(lst)

#insert
#lst.insert(index, value)

lst.insert(3,['mango',33])
print(lst)


#pop
#its remove last element
lst.pop()
print(lst)


#remove
#remove me value hatta hai
lst.remove(3)
print(lst)


#index
print(lst.index(["mango",33]))


#count
print(lst.count(2))



#sum, min,max and sorted are functions
a = [4,65,78,25,379,3,855]

#sort()
#a.sort()
#print(a)

print(min(a))
print(max(a))
print(sum(a))

print(sorted(a))
print(a)

```
