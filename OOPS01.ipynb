```python
class complex:
  def __init__(self,x,y):
    self.x=x
    self.y=y

  def __inti__(self,other):
    # return complex(self.x + other.x , self.y + other.y)
    # or
    a = self.x+other.x
    b = self.y + other.y

    result = complex(a,b)
    return result

a = complex(2,3)
b = complex(5,8)

def display(self):
  print(self.x,'+i',self.y)

c = a + b

print(c.x,c.y)
```
