```python
def grade(exam_score):
  if(exam_score>95):
    return 'A'
  elif(exam_score>70):
    return 'B'
  elif(exam_score>50):
    return 'C'
  elif(exam_score>30):
    return 'D'
  else:
    return 'F'

df['grade'] = df['exam_score'].apply(grade)
df
```
