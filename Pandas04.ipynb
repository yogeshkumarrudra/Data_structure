```python
from os import rename
df1 = pd.DataFrame({
    'id':[1,2,3,4],
    'name':['Yogesh','Rewas','Parth','sofiya'],
    'age':[23,54,21,20]
})

df2 = pd.DataFrame({
    'id':[3,4,5,6],
    'name':['Samar','Anshu','Mayank','Harshita'],
    'age':[21,40,34,19]
})
#inner join
df3 = pd.merge(df1,df2, on = 'id',how = 'inner')
display(df3)

df3.rename(columns = {'name_x':'Table_one_name1','age_x':'Table_one_age1','name_y':'Table_one_name2','name_y':'Table_one_name2'},inplace = True)
display(df3)

df4 = pd.merge(df1,df2, on = 'id',how = 'left')
display(df4)

df5 = pd.merge(df1,df2, on = 'id',how = 'right')
display(df5)

df6 = pd.merge(df1,df2, on = 'id',how = 'outer')
display(df6)

#cross wale me id nhi dayte hai q ki har ek colume dusre her ek column se intrect kar raha hai
df7 = pd.merge(df1,df2,how = 'cross')
display(df7)
```

**Concatination**


```python
df3 = pd.concat([df1,df2])
display(df3)
```
