```python
import numpy as np

stock = np.random.randint(20,80,25)
stock = np.reshape(stock,(5,5))
print(stock)

print("\n")

demand = np.random.randint(20,80,25)
demand = np.reshape(demand,(5,5))
print(demand)


print("\n")
# remanning stock

remaning_stock = stock-demand
print(remaning_stock)
print("\n")




print("\n")
print("sum_axis_0\n")
sum_axis_0=np.sum(np.where(remaning_stock<0,-1*remaning_stock,0),axis=0)
print(sum_axis_0)

print("\n")

remaning_stock= np.where(remaning_stock<0, 0,remaning_stock)
print(remaning_stock)


count_zero = np.sum(np.where(remaning_stock==0,1,0))
print(count_zero)

print("\n")
print("sort according to row wise")
remaning_stock.sort()
print(remaning_stock)

```
