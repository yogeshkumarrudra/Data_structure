```python
#change case lower(),upper(),title(),capitalize()
x="meWoWm KC"
print(x.lower())
print(x.upper())
# title only make the first letter upper case it lowers the other character after space characters are affected in the same way
print(x.title())
# capitalize only make the first letter upper case it lowers the other character and after space characters are lowerd
print(x.capitalize())
```
